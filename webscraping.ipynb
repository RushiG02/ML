```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
page = requests.get("https://www.desiblitz.com/content/bollywood-actors-actresses-height-age")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
n=[]
for a in soup.findAll('strong'):
    name=a.text
    n.append(name)
len(n)
```

```python
aa=[]
Height=np.zeros(len(n),dtype='object')
dob=np.zeros(len(n),dtype='object')
Age=np.zeros(len(n),dtype='object')
for a in soup.findAll('p'):
    aa.append(a.text)
i=0  
j=0
k=0
for x in aa:
    if 'Height' in x:
        Height[i]=x
        i+=1
    if 'Date of Birth' in x:
        dob[j]=x
        j+=1
    if 'Age' in x:
        Age[k]=x
        k+=1
```
